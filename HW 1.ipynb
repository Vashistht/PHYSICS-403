```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
import random
from numpy import sqrt, sin, cos, pi, log, histogram2d
mpl.rcParams.update(mpl.rcParamsDefault)
# plt.style.use('seaborn')
```

# Functions:
- implement the algorithm
- make 1d, 2d histograms
- contour plots

```python
# Implements the box muller algorithm with sin and cos from module 2

def box_muller(n):
    # 1. Generate n U1 and U2, which are Unif(0, 1)
    u1 = np.random.uniform(size=n)
    u2 = np.random.uniform(size=n)
    a = sqrt(-2*log(u1))
    x_1 = cos(2*pi*u2)
    y_1 = sin(2*pi*u2)
    x = a*x_1
    y = a*y_1
    return x, y
```

```python
# 1(a) - Plot scatter plot of x and y

def scatter_plot(x,y, c= "2d scatterplot"):
    plt.scatter(x,y,alpha=0.2)
    plt.title(c)
    plt.xlabel("Random Number x")
    plt.ylabel("Random Number y")
    plt.show()
```

```python
# 1(b) - Plot binned scatter plot of x and y (10 bins in both x and y)

def histogram_2D(x,y,n = 10,c = "2d histogram"):
    h = plt.hist2d(x,y,bins=(n, n), cmap=plt.cm.viridis)
    plt.title(c)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.colorbar(h[3])
    plt.show()
```

```python
# 1(c) - Plot contours of x and y (10 bins in both x and y)

def countour_plot(xs,ys, n=10, c="Contour Plot"):
    counts,ybins,xbins = histogram2d(xs,ys,bins=n)
    plt.contour(counts.transpose(),extent=[xbins[0],xbins[-1],ybins[0],ybins[-1]], cmap = plt.cm.rainbow, extend = "both")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(c)
    plt.colorbar()
    plt.show()
    plt.tight_layout()

def countourf_plot(xs,ys, n=10, c="Filled Contour Plot"):
    counts,ybins,xbins = histogram2d(xs,ys,bins=n)
    plt.contourf(counts.transpose(),extent=[xbins[0],xbins[-1],ybins[0],ybins[-1]], cmap = plt.cm.viridis, extend = "both")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(c)
    plt.colorbar()
    plt.show()
    plt.tight_layout()
```

```python
# 1(d) - Plots 1 d histogram of radial cordinate (100 bins)

def histogram_1d(x, y, bins = 100, c = "1D-histogram of r"):
    r = sqrt(x**2+y**2)
    x = r
    plt.figure()
    plt.hist(x, bins = bins)
    plt.title(c)
    plt.xlabel("r")
    plt.ylabel("frequency")
    plt.show()
```

## Plots

```python
# generate 10,000 pairs of Gaussian random numbers (x, y)
xs, ys = box_muller(10000)
```

```python
scatter_plot(xs, ys) 
```

```python
histogram_2D(xs, ys) # 10 bins is default
```

```python
histogram_1d(xs,ys)
```
